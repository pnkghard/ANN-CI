# Import the liberies

```python
import csv
import torch
import numpy
import sklearn

import time

from rdkit import Chem
from rdkit.Chem.EState import Fingerprinter
from rdkit.Chem import Descriptors
import matplotlib.pyplot as plt
```

# Data File

# Process the Training data file

```python
def fps(mol):
    return numpy.append(Fingerprinter.FingerprintMol(mol)[0],Descriptors.MolWt(mol))

with open('dataset_train.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter = ',')
    des = []
    label = []
    for row in lines:
        des.append(fps(Chem.MolFromSmiles(row[0])))
        label.append(row[4])

#print("len(des)", len(des), "len(label)", len(label) )
#print("des[:2]", des[:2], "label[:2]", labe;[:2])


# Define input features :
input_features = numpy.zeros((len(des),len(des[0])))
for i in range(len(des)):
    for j in range(len(des[0])):
            input_features[i,j] = des[i][j]

# Define target output :
target_output = numpy.zeros(len(label))
for i in range(len(label)):
    target_output[i] = label[i]


input_features = torch.tensor(input_features)
input_features = input_features.to(torch.float32)
#print ("input_features.shape", input_features.shape)
#print ("input_features",input_features)


# Reshaping our target output into vector :
target_output = target_output.reshape(len(label),1)
target_output = torch.tensor(target_output)
target_output = target_output.to(torch.float32)
#print("target_output.shape", target_output.shape)
#print ("target_output", target_output)
```

# Define the ANN structure

```python
# Define weights :
in_node = 80
h_node = 100
out_node = 1

class neural_network(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.layer1 = torch.nn.Linear(in_node, h_node)
        self.act1 = torch.nn.ReLU()
        self.layer2 = torch.nn.Linear(h_node, out_node)
        self.act2 = torch.nn.ReLU()
    def forward(self, x):
        x = self.layer1(x)
        x = self.act1(x)
        x = self.layer2(x)
        x = self.act2(x)
        return x
```

# Learning of the ANN model

```python
model = neural_network()
num_epochs = 2000
optimizer = torch.optim.Adam(model.parameters(), lr=0.001, betas=(0.9, 0.999))
loss_fn = torch.nn.MSELoss()


LOSS = []
for epoch_num in range(num_epochs):
    scores = model(input_features)
    error = loss_fn(scores, target_output)
    optimizer.zero_grad()
    error.backward()
    optimizer.step()
    print("ite: ", epoch_num, "losses: ", error.item())
    LOSS.append(error.item())

plt.plot([x for x in range(num_epochs)], LOSS)
plt.xlabel("Iteraions")
plt.ylabel("Loss (eV)")
plt.show()

train_ln = len(target_output)
print(train_ln)

scores = scores.detach().numpy()
target_output = target_output.numpy()
target_output = target_output.reshape(train_ln, 1)

train_act,  train_pred = [], []
for idx in range(train_ln):
    train_act.append(float(target_output[idx]))
    train_pred.append(float(scores[idx].item()))

plt.scatter(train_act, train_pred)
plt.show()
```

# Save  the Model

```python
torch.save(model.state_dict(), "ann_model.pth")
```
